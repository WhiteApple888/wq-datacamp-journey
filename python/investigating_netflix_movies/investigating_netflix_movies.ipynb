* Perform exploratory data analysis on the netflix_data.csv data to understand more about movies from the 1990s decade.
* What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).
* A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.
* Feel free to experiment after submitting the project!

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df_90s = netflix_df.query('release_year >= 1990 and release_year < 2000').reset_index(drop=True)
duration = int(netflix_df_90s["duration"].mode()[0])
duration
```

```python
netflix_df_90s["genre_low"] = netflix_df_90s["genre"].str.lower()
short_movie_count= int(netflix_df_90s.query('duration < 90 and genre_low == "action"').shape[0])
short_movie_count
```

* Does the historical progression of movie durations from the 1940s to the 2020s reflect a fundamental increase in typical film length, or simply a widening of creative variety within the industry?

```python
netflix_df["decade"] = (netflix_df["release_year"] // 10) * 10
decades = sorted(netflix_df["decade"].unique())
plot_duration_data = [netflix_df[netflix_df["decade"] == d]["duration"] for d in decades]
```

```python
fig, ax = plt.subplots()
ax.boxplot(plot_duration_data, labels=decades)
ax.set_title("Duration of Movies by Decades", fontsize=15)
ax.set_xlabel("Decades")
ax.set_ylabel("Duration (min)")
plt.show()
```


#### Observations
* **Expansion of Runtime Variance**: Between the 1940s and the 2020s, the box plot reveals a consistent expansion in the Interquartile Range (IQR). While movie durations were highly standardized during the 1940s and 1950s (evidenced by narrow boxes), the increasing vertical spread in recent decades suggests a shift toward more experimental and diverse runtimes.
* **Vertical Migration of the Baseline**:The box plot reveals that movies are not just becoming more varied, but fundamentally longer. We observe a consistent upward shift in the median and the interquartile range from the 1940s to the 2020s.
* **The 1960s Peak**: The 1960s represent a distinct peak in movie length; however, the long-term trend remains one of growth. 
* **Modern Elevation** Even after the 60s peak, the baseline duration (the IQR) has stayed elevated compared to the 1940s, proving that the modern 'standard' for movie length is significantly higher than it was during the 1940s.
